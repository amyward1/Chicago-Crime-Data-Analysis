# Analysis

```python
#creating column for customer sales price before discount
nw['total_sales'] = nw['quantity'] * nw['customer_unit_price']
nw
```

```python
#adding customer sales after discounts applied
nw['sale_after_discount'] = nw['total_sales'] - (nw['total_sales'] * nw['discount']) 
nw
```

```python
#Finding countries with highes sales after discounts
Top_Countries = pd.DataFrame(sqldf("SELECT shipping_country, SUM(sale_after_discount) as sales_after_discount \
FROM nw \
GROUP BY shipping_country \
Order BY sales_after_discount  DESC"))
Top_Countries
```

```python
# Countries receiving the highest percentage of discounts compared to non-discounted sales prices
Top_Country_Discounts = pd.DataFrame(sqldf("SELECT shipping_country \
,SUM(total_sales) as regular_priced_sales \
,SUM(sale_after_discount) as discounted_sales \
,SUM(sale_after_discount)/SUM(total_sales) * 100 - 100 as Percent_Discount_Received \
FROM nw \
GROUP BY shipping_country \
ORDER BY Percent_Discount_Received ASC"))
Top_Country_Discounts
```

```python
#Which categories have the highest discount rate?
Top_Category_Discounts = pd.DataFrame(sqldf("SELECT category_name \
,SUM(total_sales) as regular_priced_sales \
,SUM(sale_after_discount) as discounted_sales \
,SUM(sale_after_discount)/SUM(total_sales) * 100 - 100 as Percent_Discount_Received \
FROM nw \
GROUP BY category_name \
ORDER BY Percent_Discount_Received ASC"))
Top_Category_Discounts
```

```python
#Which individual products have the highes discount rate?
Top_Product_Discounts = pd.DataFrame(sqldf("SELECT product_name \
,category_name \
,SUM(total_sales) as regular_priced_sales \
,SUM(sale_after_discount) as discounted_sales \
,SUM(sale_after_discount)/SUM(total_sales) * 100 - 100 as Percent_Discount_Received \
FROM nw \
GROUP BY product_name \
ORDER BY Percent_Discount_Received ASC"))
Top_Product_Discounts
```
